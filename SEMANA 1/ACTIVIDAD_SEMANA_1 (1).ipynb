```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Crear un conjunto de datos
value_n_list = ['Andres Gonzalez', 'Pedro Perez', 'John Gomez', 'Maria Lopez', 'Juan Caviedes', 'Diana Arias', 'Tatiana Alvarez', 'Ricardo Lopera', 'Sara Jimenez', 'Samuel Soto']
value_matematicas_list = [4.5, 3.8, 2.0, 4.0, 2.4, 3.2, 4.8, 2.3, 1.2, 3.8]
value_español_list = [3.76, 4.87, 2.69, 3.5, 1.3, 2.8, 4.2, 4.8, 3.7, 3.9]
value_ingles_list = [3.76, 4.87, 2.69, 3.5, 4.2, 4.9, 2.2, 1.2, 4.2, 4.0]

# Crear diccionario de datos
data_dict = {
    "nombre": value_n_list,
    "matematicas": value_matematicas_list,
    "español": value_español_list,
    "ingles": value_ingles_list
}

# Convertir diccionario en DataFrame
df = pd.DataFrame(data_dict)

```

```python
df
```

```python
# Calcular el promedio por fila y crear una nueva columna
df['promedio'] = df[['matematicas', 'español', 'ingles']].mean(axis=1)

# Mostrar el DataFrame resultante
print(df)
```

```python
# Definir la función para determinar el nivel académico
def determine_nivel_academico(promedio):
    if 0 <= promedio <= 2.9:
        return "bajo"
    elif 3.0 <= promedio <= 3.9:
        return "medio"
    else:
        return "alto"

# Aplicar la función para crear la nueva columna del nivel academico
df['nivel_academico'] = df['promedio'].apply(determine_nivel_academico)

# Mostrar el DataFrame resultante
print(df)
```
